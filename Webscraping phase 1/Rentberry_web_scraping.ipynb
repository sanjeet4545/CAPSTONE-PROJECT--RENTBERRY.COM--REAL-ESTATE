```python
# Import libaraies
import pandas as pd
import numpy as np
import re

import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
import time
import re

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import warnings
warnings.filterwarnings('ignore')
```

```python
### A Function for scrape the data from rentberry website
```

```python
def scraping_data(city):                # City give city URL in argumenets.
    driver = webdriver.Chrome()
    driver.maximize_window()

    driver.get(city)

    URL = []                           
    Address = []
    Rentel_price = []
    No_of_bed = []
    No_of_bathroom = []
    Area = []
    Pet1 = []
    Pet2 = []
    Deposit = []
    Amenities = []

    start = 0
    end = 11

    for page in range(start,end):
        time.sleep(5)
        try:                 
            for i in driver.find_elements(By.XPATH,'//a[@class="property-card-bottom-address__text link-ghost font-xs one-line-string"]'):
                URL.append(i.get_attribute('href'))
        except StaleElementReferenceException:
            for i in driver.find_elements(By.XPATH,'//a[@class="property-card-bottom-address__text link-ghost font-xs one-line-string"]'):
                URL.append(i.get_attribute('href'))

        try:
            next_button = driver.find_element(By.XPATH,'//a[@class="pagination-item"][2]')
            next_button.click()
            time.sleep(3)
        except:
            next_button = driver.find_element(By.XPATH,'//a[@class="pagination-item"]')
            next_button.click()
            time.sleep(3)

    for i in URL:
        driver.get(i)
        time.sleep(3)
        # Address                 
        try:
            WebDriverWait(driver,10)
            address_tag = driver.find_element(By.XPATH,'//span[@class="property-data-address-name"]')
            Address.append(address_tag.text)
        except NoSuchElementException:
            Address.append(np.NaN)
        # Rentel_price
        try:
            WebDriverWait(driver,10)
            rentel_tag = driver.find_element(By.XPATH,'//div[@class="price-content"]/div[1]')
            Rentel_price.append(rentel_tag.text.replace('Monthly Price\n',''))
        except NoSuchElementException:
            Rentel_price.append(np.NaN)
        # No_of_bed    
        try:
            WebDriverWait(driver,10)
            bed_tag = driver.find_element(By.XPATH,'//div[@class="internals-item"][1]/div[2]')
            No_of_bed.append(bed_tag.get_attribute('textContent').replace('\n',' '))
        except NoSuchElementException:
            No_of_bed.append(np.NaN)
        # No_of_bathroom
        try:
            WebDriverWait(driver,10)
            bathroom_tag = driver.find_element(By.XPATH,'//div[@class="internals-item"][2]/div[2]')
            No_of_bathroom.append(bathroom_tag.get_attribute('textContent').replace('\n',' '))
        except NoSuchElementException:
            No_of_bathroom.append(np.NaN)
        # Area
        try:
            WebDriverWait(driver,10)
            area_tag = driver.find_element(By.XPATH,'//div[@class="internals-item"][3]/div[2]')
            Area.append(area_tag.get_attribute('textContent').replace('\n',' '))
        except NoSuchElementException:
            Area.append(np.NaN)
        # Pet1
        try:
            WebDriverWait(driver,10)
            pet1_tag = driver.find_element(By.XPATH,'//*[@class="facilities facilities-rent"]/div[4]')
            Pet1.append(pet1_tag.text.replace('\n',' '))
        except NoSuchElementException:
            Pet1.append(np.NaN)
        # Pet2  
        try:
            WebDriverWait(driver,10)
            pet2_tag = driver.find_element(By.XPATH,'//*[@class="facilities facilities-rent"]/div[5]')
            Pet2.append(pet2_tag.text.replace('\n',' '))
        except NoSuchElementException:
            Pet2.append(np.NaN)
        # Deposit    
        try:
            WebDriverWait(driver,10)
            deposit_tag = driver.find_element(By.XPATH,'//*[@class="facilities facilities-rent"]/div[6]')
            Deposit.append(deposit_tag.text.replace('\n',' '))
        except NoSuchElementException:
            Deposit.append(np.NaN)
        # Amenities
        try:
            WebDriverWait(driver,10)
            amenities_tag = driver.find_element(By.XPATH,'//div[@class="amenities-wrapper position-relative"]')
            Amenities.append(amenities_tag.text.replace('\n',',').replace('Amenities,',''))
            
        except NoSuchElementException:
            Amenities.append(np.NaN)  

    return pd.DataFrame({'Address':Address,'Rentel_price':Rentel_price,'No_of_bed':No_of_bed,
                   'No_of_bathroom':No_of_bathroom,'Area':Area,'Pet1':Pet1,'Pet2':Pet2,
                   'Deposit':Deposit,'Amenities':Amenities})
```

#### We see that some Deposit price value is going the column Pet1 and Pet2 we extract them.
